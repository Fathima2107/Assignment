# Histogram

```python
plt.figure()
plt.hist(iris['SL'])
plt.hist(iris['SW'])
plt.xlabel('SL')
plt.ylabel('SW')
iris[['SL','SW']].plot.hist()
```

# scatter plot

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(7,8))
           
plt.subplot(2,1,1)
plt.scatter(iris['SL'],iris['SW'],s=20,c='red',marker='^')
plt.title('scatter plot of SL VS SW',fontsize=15)
plt.xlabel('SL',fontsize=10)
plt.ylabel('SW',fontsize=10)
            
plt.subplot(2,1,2)
plt.scatter(iris['PL'],iris['PW'],s=20,c='green',marker='^')
plt.title('scatter plot of PL VS PW',fontsize=15)
plt.xlabel('PL',fontsize=10)
plt.ylabel('PW',fontsize=10)
           
plt.tight_layout()
```

# pair plots

```python
sns.pairplot(iris)
slice_data=iris[['SL','SW','PL','PW']]
```

# Heatmap

```python
sns.heatmap(slice_data.corr(),annot=True)
plt.xticks(rotation=90)
```
