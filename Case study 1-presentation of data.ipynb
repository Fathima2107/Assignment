# Male vs Female buyers participated in sales

```python
df=data["Buyer Gender"].value_counts().plot(kind='bar')
plt.title('Bar graph of male vs female buyers',fontsize=15)
plt.xlabel('Gender', fontsize=10)
plt.ylabel('No:of buyer gender', fontsize=10)
```

# Top 5 cars based on sales price

```python
top=data.sort_values('Sale Price',ascending=False)
print(top)
```

```python
top[["Make","Model","Sale Price"]].head()
```
