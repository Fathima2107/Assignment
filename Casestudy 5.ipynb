# 5. Check and handle outliers in at least 3 columns in the dataset

```python
plt.boxplot(data['Age'])
plt.title('Box plot of Age')
```

```python
Q1=np.percentile(data['Age'],25,interpolation='midpoint')
Q2=np.percentile(data['Age'],50,interpolation='midpoint')
Q3=np.percentile(data['Age'],75,interpolation='midpoint')
```

```python
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
```

```python
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print(low_lim)
print(up_lim)
```

```python
outlier=[]
for x in data['Age']:
    if ((x>up_lim)or(x<low_lim)):
        outlier.append(x)
data['Age']=np.where(data['Age']>up_lim,up_lim,np.where(data['Age']<low_lim,low_lim,data['Age']))
data
```

# 6. Do min max scaling on the feature set (Take ‘Survived’ as target)


```python
data.describe()
```

```python
from sklearn import preprocessing
```

```python
data.columns
```

```python
x=data.drop(['Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1)
```

```python
data= pd.get_dummies(data)
data.head()
```

```python
y=data['Survived']
x=data.drop('Survived',axis=1)
from sklearn.model_selection import train_test_split
from sklearn import linear_model
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
lr=linear_model.LinearRegression()
model=lr.fit(x_train, y_train)
from sklearn.metrics import mean_squared_error
pred=model.predict(x_test)
print('MSE is ', mean_squared_error(y_test, pred))
from sklearn.metrics import r2_score
print('R squared value is ', r2_score(y_test, pred))
```

```python
from sklearn import preprocessing
min_max = preprocessing.MinMaxScaler(feature_range=(0,2))
x = min_max.fit_transform(x)
x=pd.DataFrame(x)
x.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
lr=linear_model.LinearRegression()
model=lr.fit(x_train, y_train)
from sklearn.metrics import mean_squared_error
pred=model.predict(x_test)
print('MSE is ', mean_squared_error(y_test, pred))
from sklearn.metrics import r2_score
print('R squared value is ', r2_score(y_test, pred))
```
